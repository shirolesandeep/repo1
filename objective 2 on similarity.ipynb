# Step 1: Import Python Libraries

```python
# Data processing
import pandas as pd
import numpy as np
import scipy.stats

# Visualization
import seaborn as sns

# Similarity
from sklearn.metrics.pairwise import cosine_similarity
```

# Step 2: Read Data

```python
df=pd.read_csv('datamatrix.csv')
```

# Step 3: User Similarity (Pearson Correlation)


```python
user_similarity=df.T.corr()
```

```python
user_similarity.head()
```

```python
user_similarity.tail()
```

# Step 4:Cosine Similarity

```python
## User similarity matrix using cosine similarity
user_similarity_cosine = cosine_similarity(df.fillna(0))
user_similarity_cosine
```

# Step 5:Similarity for user 1

```python
# Pick a user ID
picked_userid = 1
# Remove picked user ID from the candidate list
user_similarity.drop(index=picked_userid, inplace=True)
# Take a look at the data
user_similarity.head()
```

```python
# Number of similar users
n = 10
# User similarity threashold
user_similarity_threshold = 0.3
# Get top n similar users
similar_users = user_similarity[user_similarity[picked_userid]>user_similarity_threshold][picked_userid].sort_values(ascending=False)[:n]
# Print out top n similar users
print(f'The similar users for user {picked_userid} are',similar_users)
```

```python
# game that the target user has played
picked_userid_gameplayed = df[df.index == picked_userid].dropna(axis=1, how='all')
picked_userid_gameplayed
```

```python
# game that similar users played. Remove game that none of the similar users have game played
similar_user_gameplayed = df[df.index.isin(similar_users.index)].dropna(axis=1, how='all')
similar_user_gameplayed
```

```python
# Remove the played game from the game list
similar_user_gameplayed.drop(picked_userid_gameplayed.columns,axis=1,inplace=True, errors='ignore')
```
